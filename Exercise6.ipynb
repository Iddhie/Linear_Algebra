## Distance between a Point and a Plane in 3D Space (1)
### Formula Proof
$Distance = \frac{|P-Q| . \vec{n}}{||\vec{n}||}$

$=Distance = \frac {Ax + By + Cz + D = 0}{||n||}$

$Ex: P \space coordinates \space (x_o, y_o, z_ o) \space and\space  Q \space coordinates  \space (x_1, y_1, z_1)$

$A(x - x1) + B(y - y1) + C(z - z1) = 0$

$\vec{n}=Ai + Bj + Ck$

Let w be the vector joining points P and Q

$Distance = |\vec{w}.\vec{n}|$

$= | (x_o - x_1, y_o - y_1, z_o - z_1). [(A, B, C)/√(A^2 + B^2 + C^2)] |$

$= |A(x_o - x_1) + B(y_o - y_1) + C(z_o - z_1)|/√(A^2 + B^2 + C^2)$

$= | Ax_o + By_o + Cz_o - (Ax_1 + By_1 + Cz_1) |/√(A^2 + B^2 + C^2)$

$= | Ax_o + By_o + Cz_o + D |/√(A^2 + B^2 + C^2)$

$=Distance = \frac{|P-Q| . \vec{n}}{||\vec{n}||}$

## Distance between a Point and a Plane in 3D Space (2)

$Distance = \frac{||(2,2,2).(2,-1,4)||}{\sqrt{21}}$

$Distance=\frac{10){4.58258} = 2.182$

=2.182

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

P = np.array([3, 4, 5])
Q = np.array([1, 2, 3])
normal = np.array([2, -1, 4])

A, B, C = normal[0], normal[1], normal[2]
p, q, r = Q[0], Q[1], Q[2]


def plane_equation(A, B, C, p, q, r, x, y):
    return (-A * x - B * y + A * p + B * q + C * r) / C

# Calculate the distance
distance = np.abs(np.dot(P - Q, normal)) / np.linalg.norm(normal)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(P[0], P[1], P[2], c='r', marker='o', label='Point P')
ax.scatter(Q[0], Q[1], Q[2], c='g', marker='o', label='Point Q')

xx, yy = np.meshgrid(range(15), range(15))
zz = plane_equation(A, B, C, p, q, r, xx, yy)
ax.plot_surface(xx, yy, zz, alpha=0.5)

perpendicular_vector = np.array(normal) * distance
ax.quiver(P[0], P[1], P[2], perpendicular_vector[0], perpendicular_vector[1], perpendicular_vector[2], color='b', label='Perpendicular from P')

ax.quiver(0, 0, 0, normal[0], normal[1], normal[2], color='y', label='Normal vector')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Distance from Point to Plane')

ax.text(2, 3, 6, f'Distance = {distance:.2f}', fontsize=12, color='b')

plt.legend()
plt.show()
```

## Distance between two Parallel Planes (1)

$Distance = \frac {|D_1-D_2|}{√(A^2 + B^2 + C^2)}$

$A=3, B=-4, C=2, D_1=-7, D_2=8$

$Distance = \frac {|-15|}{√(3^2 + (-4)^2 + 2^2)}$

$Dsitance =\frac{15}{5.3852}=2.7854$



```python
import numpy as np
import matplotlib.pyplot as plt

A, B, C = 3, -4, 2
D1, D2 = -7, 8

distance = np.abs(D1 - D2) / np.sqrt(A**2 + B**2 + C**2)

print(f"The distance between the two parallel planes is: {distance}")

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x_values = np.linspace(-10, 10, 100)
y_values = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x_values, y_values)

z1 = (D1 - A*x - B*y) / C
z2 = (D2 - A*x - B*y) / C

ax.plot_surface(x, y, z1, alpha=0.5)

ax.plot_surface(x, y, z2, alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Two Parallel Planes in 3D Space')

plt.show()
```

## Distance from a Plane to a Point when the Plane Equation is given (1)
### Fromula Proof
$Distance = \frac {Ax1 + By1 + Cz1 - D}{√(A^2 + B^2 + C^2)}$

Consider a point P with coordinates $(x1, y1, z1)$

Normal Vector $\vec {n}=(A,B,C)$

Point Q with coordinates $(x2, y2, z2)$ on the plane

Then, the equation can be rewritten as $Ax + By + Cz + (- Ax2 - By2 - Cz2) = 0$

$Ax + By + Cz + D = 0,\space where \space D = - (Ax2+ By2 + Cz2)$

Let w be the vector joining points P and Q.

$w=(x1 - x2, y1 - y2, z1 - z2)$

The distance between point P and the given plane is the length of the projection of vector w onto the unit normal vector n

$Distance = |\vec{w}.\vec{n}|$

$= | (x_1 - x_2, y_1 - y_2, z_1 - z_2). [(A, B, C)/√(A^2 + B^2 + C^2)] |$

$= |A(x_1 - x_2) + B(y_1 - y_2) + C(z_1 - z_2)|/√(A^2 + B^2 + C^2)$

$= | Ax_1 + By_1 + Cz_1 - (Ax_2 + By_2 + Cz_2) |/√(A^2 + B^2 + C^2)$

$= | Ax_1 + By_1 + Cz_1 - D |/√(A^2 + B^2 + C^2)$

## Distance from a Plane to a Point when the Plane Equation is given (2)

$A=1,B=2,C=3, D=12$

Point $P=(-5,1,2)$

$Distance =\frac{| -5 + 2 + 6 - 12 |}{√(1^2 + 2^2 + 3^2)}$

$Distance =\frac{9}{3.74166}=2.4053$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

n = np.array([1, 2, 3])
d = 12
P = np.array([-5, 1, 2])

def plane_equation(x, y, d, n):
    return (d - n[0]*x - n[1]*y) / n[2]

distance = abs(np.dot(n, P) - d) / np.linalg.norm(n)

print(f"The distance between the point {P} and the plane is: {distance}")

Q = P - (np.dot(n, P) - d) / np.dot(n, n) * n
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x_values = np.linspace(-10, 10, 100)
y_values = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x_values, y_values)
z = plane_equation(x, y, d, n)

ax.plot_surface(x, y, z, alpha=0.5)

ax.scatter(P[0], P[1], P[2], color='r', s=20)

ax.plot([P[0], Q[0]], [P[1], Q[1]], [P[2], Q[2]], color='g')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Point, Plane, and Line in 3D Space')

plt.show()
```

## Angle between Planes (1)
### Proof
The normals of the planes indicates the direction which the plane is faced.

The dot product between N1 and N2:

$N1 · N2 = |N1| * |N2| * cos(θ)$

$cos(θ) = (N1 · N2) / (|N1| * |N2|)$

$θ=arrcos((N1 · N2) / (|N1| * |N2|))$

## Angle between Planes (2)
Planes: $3x+4y-5z=11, 3x+4y-5z=-16$

$cos(θ)=\frac{50}/{7.07106*7.07106}$

$=1$

$θ=cos^ (1)=0 \space degrees$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the plane equations
A1, B1, C1, D1 = 3, 4, -5, 11
A2, B2, C2, D2 = 3, 4, -5, -16

n1 = np.array([A1, B1, C1])
n2 = np.array([A2, B2, C2])

cos_theta = np.abs(np.dot(n1, n2)) / (np.linalg.norm(n1) * np.linalg.norm(n2))
theta = np.arccos(cos_theta)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

Z1 = (D1 - A1 * X - B1 * Y) / C1
Z2 = (D2 - A2 * X - B2 * Y) / C2

ax.plot_surface(X, Y, Z1, alpha=0.5)
ax.plot_surface(X, Y, Z2, alpha=0.5)

origin = (0, 0, 0)
ax.quiver(origin[0], origin[1], origin[2], n1[0], n1[1], n1[2], color='r')
ax.quiver(origin[0], origin[1], origin[2], n2[0], n2[1], n2[2], color='b')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Angle Between Two Planes in 3D Space')

plt.show()

print(f"The angle between the planes is: {np.degrees(theta)} degrees")
```
