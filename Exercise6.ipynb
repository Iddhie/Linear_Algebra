```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a = np.array([4,2,3])
d = np.array([6, 3, 1])
p = np.array([5, 1, 7])

# Calculate the distance between the point and the line
distance_cross = np.linalg.norm(np.cross(d, p - a)) / np.linalg.norm(d)

# Calculate the distance between the point and the line using the projection of (u-b) ont to v
distance_dot = np.linalg.norm((p-a) - np.dot(d, p - a)/ np.dot(d,d) * d)

print(f"The distance between the point and the line using the cross product formula is: {distance_cross}")
print(f"The distance between the point and the line using the dot product formula is: {distance_dot}")

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the line
t = np.linspace(0, 1, 100)
line = np.array([a + ti * d for ti in t])
ax.plot(line[:, 0], line[:, 1], line[:, 2], label='Line')


ax.scatter(p[0], p[1], p[2], color='r', label='Point')

# Plot the connecting line between the point and the closest point on the line
closest_point = a + np.dot(p - a, d) / np.dot(d, d) * d
ax.plot([p[0], closest_point[0]], [p[1], closest_point[1]], [p[2], closest_point[2]], color='g', label='Distance')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Distance Between Point and Line in 3D Space')
ax.legend()

# Show the plot
plt.show()
```

## Equation of a Plane 1

###Plane Equation  : 2x - 4y + 3z = 22

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

n = np.array([2, -4, 3])
Q = np.array([1, -2, 4])

A, B, C = n[0], n[1], n[2]
x_0, y_0, z_0 = Q[0], Q[1], Q[2]

# Define the equation of the plane
def plane_equation(x, y):
    return (-A * x - B * y + A * x_0 + B * y_0 + C * z_0) / C

x_values = np.linspace(-20, 20, 100)
y_values = np.linspace(-20, 20, 100)
x, y = np.meshgrid(x_values, y_values)
z = plane_equation(x, y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(x, y, z, alpha=0.5)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plane in Point-Normal Form')

# Show the plot
plt.show()
```

## Equation of a Plane 3

1. Represents the vector components of the normal of the plane. The normal indicates the direction which the plane's surface is pointing

2.Determines the distance from the origin to the plane when taken from the direction of the normal vector.

3.Euclidean norm, also can be called magnitude is the length of the normal vector. It is used as a scaling factor to normalize the distance ensuring  that the distance calculated is not influenced by the magnitudes of the coefficients.

4.The plane passes through the origin

5.$$Ax + By + Cz = D_1,$$
$$Ax + By + Cz = D_2.$$
These planes are parallel to each other and have the same orientation in space. The distance between the two parallel planes is given by
$$\text{Distance} = \frac{|D_2 - D_1|}{\sqrt{A^2 + B^2 + C^2}}.$$
This calculation indicates how far apart they are along the direction of the normal vector.

## Converting a Plane Equation from Standard Form to Parametric Form (1)
The plane equation is $3x-4y+2z=5$.

Point on the plane  = $(0,0,2.5)$

$z = 2.5 - 1.5s + 2t$

$(s,t,2.5-1.5s+2t)$


We can now split this into two vectors,

$s(1,0,-1.5)+t(0,1,2)$


$P(s,t)=(0,0,2.5)+s(1,0,-1.5)+t(0,1,2)$


We can check for correctness by taking an example point from the parametric equation: Let s=1 and t=1: P(s=1,t=1)=(0,0,-2.5)+(1,0,-1.5)+(0,1,2)=(1,1,1)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parameters for the plane
A, B, C, D = 3, -4, 2, 5

# Define the point on the plane and the two vectors
P0 = np.array([0, 0, 2.5])
v = np.array([1, 0, -1.5])
w = np.array([0, 1, 2])

# Define the parametric equations
def plane_parametric(u, v):
    return P0 + u * v + v * v

# Generate points for the plane
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
x, y = np.meshgrid(x, y)
z = (D - A * x - B * y) / C

# Create the figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(x, y, z, alpha=0.5)

# Plot the point and vectors
ax.quiver(P0[0], P0[1], P0[2], v[0], v[1], v[2], color='r')
ax.quiver(P0[0], P0[1], P0[2], w[0], w[1], w[2], color='g')
ax.scatter(P0[0], P0[1], P0[2], color='b')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plane and Vectors')

# Show the plot
plt.show()
```

## Converting a Plane Equation from Parametric Form to Standard Form (1)

$P_0 = (1,0,-2)$ , $v=(1,3,6), w=(4,-7,2)$

$v * w = (48,22,-19) = n$

$ n . P_0 = (1,0,-2) . (48,22,-19) = 86$

$48x + 22y -19z =86$

For correctness lets use u = 1 and v = 1

$P_(1,1)=(1,0,-2)+1(1,3,6)+1(4,-7,2)=(6,-4,6)$

$48*6+22*(-4)-19*6 =86$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
P0 = np.array([1, 0, -2])
v = np.array([1, 3, 6])
w = np.array([4, -7, 2])

# Find the cross product to get the normal vector
n = np.cross(v, w)

# Define the standard form parameters
A, B, C = n[0], n[1], n[2]
D = np.dot(n, P0)

# Generate points for the plane
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
x, y = np.meshgrid(x, y)
z_standard = (D - A * x - B * y) / C

# Create the figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane in standard form
ax.plot_surface(x, y, z_standard, alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plane in Standard Form')

# Show the plot
plt.show()
```

## Finding the Plane Equation from 3 Points (1)
Points are $Q(1, 2, 3)$, $R(5,6,2)$, and $S(7, 11, 4)$.

The vectors $\mathbf{QR}$ and $\mathbf{QS}$ are then:

$\vec{QR} = (5 - 1, 6 - 2, 2 - 3) = (4, 4, -1)$

$\vec{QS} = (7 - 1, 11 - 2, 4 - 3) = (6, 9, 1)$

$QR * QS = (13,-10,12)$

$ \vec{\mathbf{QP}} = (x - 1, y - 2, z - 3)$

For a point to lie in the plane, the dot product of $\mathbf{v}$ and $\mathbf{n}$ must be zero:

$\vec{n} \cdot \vec{\mathbf{QP}}  = 13(x-1) - 10(y - 2) + 12(z - 3) = 0$

$13x - 13 - 10y + 20 + 12z - 36 = 0$

$13x - 10y + 12z = 29$

Thus, the equation of the plane passing through the points $Q$, $R$, and $S$ is  $$13x - 10y + 12z = 29$$

Check for correctness:
Check if point R (7,11,4) is on the plane.
$$13x - 10y + 12z \quad \text{should equal} \quad 29$$
$$13 \times 7 - 10 \times 11 + 12 \times 4 = 29$$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the points Q, R, and S
Q = np.array([1, 2, 3])
R = np.array([5, 6, 2])
S = np.array([7, 11, 4])

# Calculate the vectors QR and QS
QR = R - Q
QS = S - Q

# Calculate the cross product of QR and QS to get the normal vector
n = np.cross(QR, QS)

# Define the parameters for the standard form
A, B, C = n
D = np.dot(n, Q)

# Generate the standard form of the plane equation
def plane_standard(A, B, C, D, x, y):
    return (D - A * x - B * y) / C

# Create a grid for the plane
x = np.linspace(-5, 15, 20)
y = np.linspace(-5, 15, 20)
x, y = np.meshgrid(x, y)
z = plane_standard(A, B, C, D, x, y)

# Create the figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(x, y, z, alpha=0.5)

# Mark the points Q, R, S
ax.scatter(Q[0], Q[1], Q[2], color='r', s=10, label='Q')
ax.scatter(R[0], R[1], R[2], color='g', s=10, label='R')
ax.scatter(S[0], S[1], S[2], color='b', s=10, label='S')

# Set the labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plane with Points Q, R, and S')

# Show the plot
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the points Q, R, and S
Q = np.array([11, 4, 5])
R = np.array([3, 1, 2])
S = np.array([5, 3, 4])

# Calculate the vectors QR and QS
QR = R - Q
QS = S - Q

# Calculate the cross product of QR and QS to get the normal vector
n = np.cross(QR, QS)

# Define the parameters for the standard form
A, B, C = n
D = np.dot(n, Q)

# Generate the standard form of the plane equation
def plane_standard(A, B, C, D, x, y):
    return (D - A * x - B * y) / C

# Create a grid for the plane
x = np.linspace(-5, 15, 20)
y = np.linspace(-5, 15, 20)
x, y = np.meshgrid(x, y)
z = plane_standard(A, B, C, D, x, y)

# Create the figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(x, y, z, alpha=0.5)

# Mark the points Q, R, S
ax.scatter(Q[0], Q[1], Q[2], color='r', s=10, label='Q')
ax.scatter(R[0], R[1], R[2], color='g', s=10, label='R')
ax.scatter(S[0], S[1], S[2], color='b', s=10, label='S')

# Set the labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plane with Points Q, R, and S')

# Show the plot
plt.legend()
plt.show()
```

$\vec{QS} = (-5 - (-6), 1 - 14, 6 - 7) = (1, -13, -1)$

$QR * QS = (13,8,-91)$

$ \vec{\mathbf{QP}} = (x + 6, y - 14, z - 7)$

$\vec{n} \cdot \vec{\mathbf{QP}}  = 13(x-11) +8(y - 4) - 91(z - 5) = 0$

Thus, the equation of the plane passing through the points $Q$, $R$, and $S$ is  $$13x+8y - 91z = 380$$
