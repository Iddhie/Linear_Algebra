```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Magnitude of the Vectors

```python
v1=np.array([3,4,2])
v2=np.array([1,2,5])

# Calculate the magnitude of V1
mag_manual = np.sqrt(np.sum(v1 ** 2))  # Manual calculation of the magnitude
mag_numpy = np.linalg.norm(v1)  # Numpy calculation of the magnitude

# Calculate the magnitude of V2
mag2_manual = np.sqrt(np.sum(v2 ** 2))  # Manual calculation of the magnitude
mag2_numpy = np.linalg.norm(v2)  # Numpy calculation of the magnitude

print(mag_manual)
print(mag_numpy)
print(mag2_manual)
print(mag2_numpy)
```

#### Direction of Vectors

```python
# Calculate the direction of the vector
direction1 = v1 / mag_manual
direction2 = v2/mag2_manual

print(direction1)
print(direction2)
```

### Plotting the Vectors using Matplotlib

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vector v as an arrow
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='Vector 1 (V)', linewidth=2)

# Plot vector u as an arrow
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', label='Vector 2 (U)', linewidth=2)

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set axis limits
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_zlim(0, 5)

# Add a legend
ax.legend()

# Show the plot
plt.grid()
plt.show()

```

### Plotting the resultant vectors of the linear combinations

```python
# Calculate the resulting vectors
v_sub_u = v1 - v2
u_sub_v = v2 - v1
u_add_v = v2 + v1

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vector v as an arrow
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='Vector 1 (V)', linewidth=2)

# Plot vector u as an arrow
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', label='Vecror 2 (U)', linewidth=2)

# Plot the resulting vector v - u as an arrow
ax.quiver(0, 0, 0, v_sub_u[0], v_sub_u[1], v_sub_u[2], color='g', label='v - u', linewidth=2)

# Plot the resulting vector u - v as an arrow
ax.quiver(0, 0, 0, u_sub_v[0], u_sub_v[1], u_sub_v[2], color='c', label='u - v', linewidth=2)

# Plot the resulting vector u + v as an arrow
ax.quiver(0, 0, 0, u_add_v[0], u_add_v[1], u_add_v[2], color='m', label='u + v', linewidth=2)

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set axis limits
ax.set_xlim(-5, 10)
ax.set_ylim(-5, 10)
ax.set_zlim(-5, 10)

# Add a legend
ax.legend()

# Show the plot
plt.grid()
plt.show()
```
