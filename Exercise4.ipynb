## Vector Decomposition and Visualization (1)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors u and w
u = np.array([3, 4])  # Example vector
w = np.array([1, 2])  # Reference vector

# Calculate the projection of u onto w
proj_w_u = np.dot(u, w) / np.dot(w, w) * w
print(proj_w_u)

# Calculate the orthogonal component of u with respect to w
ortho_w_u = u - proj_w_u
print(ortho_w_u)

# Plot the vectors u, w, proj_w_u, and ortho_w_u
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u')
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='g', label='w')
plt.quiver(0, 0, proj_w_u[0], proj_w_u[1], angles='xy', scale_units='xy', scale=1, color='b', alpha = 0.2, label='proj_w_u')
plt.quiver(proj_w_u[0], proj_w_u[1], ortho_w_u[0], ortho_w_u[1], angles='xy', scale_units='xy', scale=1, color='c', alpha = 0.2, label='ortho_w_u')

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5, alpha = 0.7)
plt.legend()
plt.show()

```
