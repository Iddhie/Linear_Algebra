## Equation of a Line (1)
_________________________

P(7, 4, -1) and a vector v = (1, 3, 4)

$$x(t) = t + 7$$
$$y(t) = 3t + 4$$
$$z(t) = 4t - 1$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parametric equations for the line
t = np.linspace(-10, 10, 100)  # Range of t values
x = t + 7
y = 3*t + 4
z = 4*t - 1

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the line
ax.plot(x, y, z, label='Line', color='b')

# Plot the point P
ax.scatter(7, 4, -1, color='r', label='Point P(7, 4, -1)')

# Set labels and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.legend()

# Show the plot
plt.show()
```

## Equation of a Line (2)
_________________________
Line 1:
$$\frac{x - 2}{2} = \frac{y - 3}{1} = \frac{z - 1}{4}$$

Line 2:
$$\frac{x - 4}{3} = \frac{y - 2}{2} = \frac{z - 3}{5}$$

For Line 1:

$$x = 2 + 2t$$

$$y = 3 + t$$

$$z = 1 + 4t$$

For Line 2:

$$x = 4 + 3s$$

$$y = 2 + 2s$$

$$z = 3 + 5s$$

For $x$:

$$2 + 2t = 4 + 3s$$

$$ : 2t - 3s = 2$$

For $y$:

$$3 + t = 2 + 2s$$

$$ : t-2s = -1 $$

Eliminate t,

$$ 2t -3s = 2 $$

$$ 2t - 4s = -2 $$

$$ : s = 4 $$


$$ t - 2(4) = -1 $$

$$ t = 7 $$

#### Find x, y, z values for P by substituting t = 7,

$$ z = 1 + 4(7) = 29 $$

$$ x = 2 + 2(7) = 16 $$

$$ y = 3 + 7 = 10 $$

Therefore, P = (16, 10, 29)



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parametric equations for the two lines
def line1(t):
    x = 2 + 2 * t
    y = 3 + t
    z = 1 + 4*t
    return x, y, z

def line2(s):
    x = 4 + 3 * s
    y = 2 + 2 * s
    z = 3 + 5 * s
    return x, y, z

# Define the range of t and s values
t_values = np.linspace(-20, 0, 100)
s_values = np.linspace(-20, 0, 100)

# Calculate the points for the two lines
x1, y1, z1 = line1(t_values)
x2, y2, z2 = line2(s_values)

intersection_point = (16, 10, 29)

# Plot the lines and the intersection point
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x1, y1, z1, label='Line 1')
ax.plot(x2, y2, z2, label='Line 2')
ax.scatter(*intersection_point, color='r', label='Intersection')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.title('Intersection of Two Lines')
plt.legend()
plt.show()
```

## Equation of a Line (3)
_________________________

For Line 1:

$$x = 6 + 3t$$

$$y = 5 + 4$$

$$z = 7 + 2t$$

For Line 2:

$$x = 11 + 4s$$

$$y = 3 + 3s$$

$$z = 1 + 5s$$

For $x$:

$$6 + 3t = 11 + 4s$$

$$ : 3t - 4s = 5$$

For $y$:

$$5 + 4t = 3 + 3s$$

$$ : 4t-3s = -2 $$

Eliminate t,

$$ 12t -16s = 20 $$

$$ 12t - 9s = -6 $$

$$ : s = \frac {-26}{7} $$


$$ 3t - 4(s) = 5 $$

$$ t = \frac {-23}{7}$$

#### Find x, y, z values for P by substituting t = 7,

$$ x = 6 + 3(\frac {-23}{7}) = -3.8571 $$

$$ y = 5 + 4(\frac {-23}{7}) = -8.1429 $$

$$ z = 7 + 2(\frac {-23}{7}) = 0.4286 $$

Therefore, P = (16, 10, 29)



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parametric equations for the two lines
def line1(t):
    x = 6 + 3 * t
    y = 5 + 4*t
    z = 7 + 2*t
    return x, y, z

def line2(s):
    x = 11 + 4 * s
    y = 3 + 3 * s
    z = 1 + 5 * s
    return x, y, z

# Define the range of t and s values
t_values = np.linspace(-20, 0, 100)
s_values = np.linspace(-20, 0, 100)

# Calculate the points for the two lines
x1, y1, z1 = line1(t_values)
x2, y2, z2 = line2(s_values)

intersection_point = (-3.8571 , -8.1429, 0.4286)

# Plot the lines and the intersection point
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x1, y1, z1, label='Line 1')
ax.plot(x2, y2, z2, label='Line 2')
ax.scatter(*intersection_point, color='r', label='Intersection')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.title('Intersection of Two Lines')
plt.legend()
plt.show()
```

## Exercise: Equation of a Line (4)


```python
import numpy as np

# Define Line 1 parameters
a1 = np.array([1, 2, 3, 4, 5])
d1 = np.array([2, 3, 1, 4, 2])

# Define Line 2 parameters
a2 = np.array([7, -4, 9, 14, 1])
d2 = np.array([2, 3, 1, 4, 2])

# Calculate the vector connecting the two lines
v = a2 - a1

# Calculate the unit vector along Line 1
u1 = d1 / np.linalg.norm(d1)

# Project v onto u1
projection = np.dot(v, u1) * u1

# Calculate the shortest distance
shortest_distance = np.linalg.norm(projection)

print(f"The shortest distance between Line 1 and Line 2 is: {shortest_distance:.2f}")
```
