## Determinant 1

```python
import matplotlib.pyplot as plt
import numpy as np
def calculate_determinant(matrix):
    # Extract matrix elements
    a11, a12 = matrix[0][0], matrix[0][1]
    a21, a22 = matrix[1][0], matrix[1][1]

    # Calculate determinant
    determinant = a11 * a22 - a12 * a21

     # Create column vectors
    v1 = np.array([a11, a21])
    v2 = np.array([a12, a22])

    # Create a figure
    fig, ax = plt.subplots()

    # Plot the vectors
    ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
    ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')

    # Set plot limits
    ax.set_xlim([0, 10])
    ax.set_ylim([0, 12])

    # Set aspect ratio of the plot to be equal
    ax.set_aspect('equal')

    # Add labels and legend
    ax.text(v1[0] / 2, v1[1] / 2, f'Det(A) = {determinant:.2f}', ha='center', va='center', fontsize=12, color='green')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend()

    # Show the plot
    plt.grid()
    plt.show()

    return determinant

matrix_A = [
    [-11, 5],
    [2, 2]
]

determinant_A = calculate_determinant(matrix_A)
print("Determinant of matrix A:", determinant_A)

matrix_B=[
    [4,6],
    [8,5]
]
determinant_B = calculate_determinant(matrix_B)
print("Determinant of matrix A:", determinant_B)
```

## Determinant 2

```python
def calculate_determinant_3x3(matrix):
    # Extract matrix elements
    a11, a12, a13 = matrix[0][0], matrix[0][1], matrix[0][2]
    a21, a22, a23 = matrix[1][0], matrix[1][1], matrix[1][2]
    a31, a32, a33 = matrix[2][0], matrix[2][1], matrix[2][2]

    # Calculate determinant
    determinant = a11 * (a22 * a33 - a23 * a32) - a12 * (a21 * a33 - a23 * a31) + a13 * (a21 * a32 - a22 * a31)

    return determinant

matrix_A = [
    [-3, 5, 3],
    [4, 5, 6],
    [-7, 8, 7]
]

determinant_A = calculate_determinant_3x3(matrix_A)
print("Determinant of matrix A:", determinant_A)

matrix_B = [
    [1, 2, 3],
    [100, 4, 9],
    [-8, 23, -4]
]

determinant_B = calculate_determinant_3x3(matrix_B)
print("Determinant of matrix A:", determinant_B)
```

## Inverse of a 2x2 Matrix (1)

$$ A = \begin{bmatrix}
3 & 5 \\
2 & 7
\end{bmatrix} $$

1. Calculate the determinant of matrix $ A $ using the formula: $$ \text{det}(A) = 3*7 - 5*2 =11$$

3. Calculate the cofactors $ C_{ij} $ for each element $ a_{ij} $ using the following formulas:
   $$ C_{11} = 3 $$
   $$ C_{12} = -5 $$
   $$ C_{21} = -2 $$
   $$ C_{22} = 7 $$

4. Construct the cofactor matrix:
   $$ \text{Cofactor Matrix} = \begin{bmatrix}
   C_{11} & C_{12} \\
   C_{21} & C_{22}
   \end{bmatrix} = \begin{bmatrix}
   3 & -5 \\
   -2 & 7
   \end{bmatrix} $$

5. Take the transpose of the cofactor matrix to obtain the adjugate matrix $ \text{adj}(A) $:
   $$ \text{adj}(A) = \text{Cofactor Matrix}^T = \begin{bmatrix}
   7 & -5 \\
   -2 & 3
   \end{bmatrix} $$

6. Finally, find the inverse matrix $ A^{-1} $ using the formula:
   $$ A^{-1} = \frac{1}{11} \cdot \text{adj}(A) $$

$$ A^{-1} = \begin{bmatrix}
7/11 & -5/11 \\
-2/11 & 3/11
\end{bmatrix} $$

## Inverse of a 2x2 Matrix (2)
$$ A = \begin{bmatrix}
-11 & 5 \\
2 & 2
\end{bmatrix} $$

1. Calculate the determinant of matrix $ A $ using the formula: $$ \text{det}(A) = -11*2 - 5*2 =32$$

3. Calculate the cofactors $ C_{ij} $ for each element $ a_{ij} $ using the following formulas:
   $$ C_{11} = -11 $$
   $$ C_{12} = -5 $$
   $$ C_{21} = -2 $$
   $$ C_{22} = 2 $$

4. Construct the cofactor matrix:
   $$ \text{Cofactor Matrix} = \begin{bmatrix}
   C_{11} & C_{12} \\
   C_{21} & C_{22}
   \end{bmatrix} = \begin{bmatrix}
   -11 & -5 \\
   -2 & 2
   \end{bmatrix} $$

5. Take the transpose of the cofactor matrix to obtain the adjugate matrix $ \text{adj}(A) $:
   $$ \text{adj}(A) = \text{Cofactor Matrix}^T = \begin{bmatrix}
   2 & -5 \\
   -2 & -11
   \end{bmatrix} $$

6. Finally, find the inverse matrix $ A^{-1} $ using the formula:
   $$ A^{-1} = \frac{1}{-32} \cdot \text{adj}(A) $$

$$ A^{-1} = \begin{bmatrix}
-1/16 & 5/32 \\
1/16 & 11/32
\end{bmatrix} $$

## Analyzing Rank and Null Space (1)
$$ A = \begin{bmatrix} -3 & 5 & 3 \\ 4 & 5 & 6 \\ -7 & 8 & 7 \end{bmatrix} $$

1. $$ \text{det}(A) = 110$$ Non Singular

2. Rank = $3$ This is a full rank

3. All rows are linearly independent

4. $$ N(A) = \begin{bmatrix} 0 \\ 0\\ 0 \end{bmatrix} $$

5. $$ A^{-1} = \begin{bmatrix} 13/110 & 1/10 & -3/22 \\ 7/11 & 0 & -3/11 \\ -67/110 & 1/10 & 7/22 \end{bmatrix} $$
The determinant is nonzero and the matrix is square, thus the matrix is non-singular, and its inverse exists. And can be calculated using the formala below:
 $$ A^{-1} = \frac{1}{\text{det}(A)} \cdot \text{adj}(A) $$

6. A matrix is singular if its determinant is zero, indicating linear dependence among its columns and the absence of an inverse. In contrast, non-singular matrices with full rank have linearly independent columns, ensuring unique solutions to linear systems and invertibility. The inverse of a non-singular matrix allows for the retrieval of the original matrix, and its full rank implies a trivial null space, crucial for one-to-one transformations. The importance of full rank lies in guaranteeing uniqueness, computational stability, and well-posed linear systems, making it a fundamental concept in applications such as optimization, machine learning, and numerical computations.
