### Calculating Dot Product

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Define the vectors
v = np.array([3, 4, 2])
w = np.array([1, 2, 5])

# Calculate the dot product of v and w
dot_vw = np.dot(v, w)

dot_vw
```

### Calculating angle between the vectors and verifying dot product

```python
# Calculate the magnitudes of v and w
magv = np.linalg.norm(v)
magw = np.linalg.norm(w)

# Calculate the angle θ between the two vectors using the inverse cos
cos_theta = dot_vw / (magv * magw)
theta = np.arccos(cos_theta)

print("Angle θ (in radians):", theta)
print(magv*magw*np.cos(theta))
```

### Visualizing the 3D Vectors and the Angle between them

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vector v as an arrow
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='r', label='v', linewidth=2)

# Plot vector w as an arrow
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='b', label='w', linewidth=2)

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set axis limits
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_zlim(0, 5)

# Add a legend
ax.legend()

# Add a text label for the angle θ
ax.text(0.5, 0.5, 1.25, f'θ = {np.degrees(theta):.2f}°', fontsize=12, color='k')

# Show the plot
plt.title('Vectors v and w with Angle θ')
plt.grid()
plt.show()
```
