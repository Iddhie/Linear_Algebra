### Calculating Cross Product

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Define the two 3D vectors
v = np.array([3, 4, 2])
w = np.array([1, 2, 5])

# Calculate the cross product manually
cross_manual = np.array([v[1]*w[2] - v[2]*w[1], v[2]*w[0] - v[0]*w[2], v[0]*w[1] - v[1]*w[0]])

# Calculate the cross product using NumPy's built-in function
cross_numpy = np.cross(v, w)

print(cross_manual)
print(cross_numpy)
```

### Calculating Cross Product Magnitude and Visualising

```python
# Calculate magnitude of the cross product
magcross = np.linalg.norm(cross_numpy)
print(magcross)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the two vectors
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='b', label='Vector v')
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='r', label='Vector w')

# Plot the cross product
ax.quiver(0, 0, 0, cross_numpy[0], cross_numpy[1], cross_numpy[2], color='g', label='Cross Product')

# Define the points for the parallelogram
points = np.array([[0,0,0], [0,0,0], v, w,[0,0,0]])

# Create a polygon collection
polygon = Poly3DCollection([points])
polygon.set_color('#90EE90')
polygon.set_edgecolor('k')
polygon.set_alpha(0.9)
ax.add_collection3d(polygon)

# Calculate and plot the area of the parallelogram
area = magcross
ax.text(3, 3, 6, f'Area: {area}', color='k')

# Set limits for the plot
ax.set_xlim([0, 20])
ax.set_ylim([-10, 10])
ax.set_zlim([0, 10])

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

### Calculating Vectors AB and AC

```python
A = np.array([2, 3, 1])
B = np.array([4, 1, 2])
C = np.array([5, 3, 6])

AB = B-A
AC = C-A
print(AB)
print(AC)
```

```python
crossabac=np.array([AB[1]*AC[2] - AB[2]*AC[1], AB[2]*AC[0] - AB[0]*AC[2], AB[0]*AC[1] - AB[1]*AC[0]])
crossabac
```

```python
magN=np.linalg.norm(crossabac)
magAB = np.linalg.norm(AB)
magAC =np.linalg.norm(AC)

print(magN)

#Area of the triangle:
Triangle_Area=magN/2
print("Area of the triange:",Triangle_Area)
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot points A, B, and C
ax.scatter(*A, color='r', label='A')
ax.scatter(*B, color='g', label='B')
ax.scatter(*C, color='b', label='C')

# Plot vectors AB and AC
ax.quiver(*A, *AB, color='k', label='AB', linewidth=2)
ax.quiver(*A, *AC, color='m', label='AC', linewidth=2)

# Set axis limits
ax.set_xlim(0, 6)
ax.set_ylim(0, 6)
ax.set_zlim(0, 6)

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a legend
ax.legend()

# Show the plot
plt.title('Triangle ABC with Vectors AB and AC')
plt.grid()
plt.show()
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the vertices of the triangle
x = [A[0], B[0], C[0], A[0]]
y = [A[1], B[1], C[1], A[1]]
z = [A[2], B[2], C[2], A[2]]

# Plot the triangle
ax.plot_trisurf(x, y, z, linewidth=0.2, antialiased=True, color="b", alpha=0.5)

# Set labels for each axis
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Show the plot
plt.show()
```
